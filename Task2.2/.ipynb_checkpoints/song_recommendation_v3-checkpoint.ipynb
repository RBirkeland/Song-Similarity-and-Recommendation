```python
# IMPORTS
import numpy as np
import time
import math
import pandas as pd
import collections
import matplotlib.pyplot as plt
import time
from scipy.sparse import coo_matrix
from scipy.sparse import csr_matrix
from scipy.sparse import lil_matrix
from scipy.sparse.linalg import svds
import random
%matplotlib inline
```

# Helper Functions

```python
def hash_count(n, b):
    
    n = float(n)
    largest = int(math.log(n, 2)) #largest power of two < n
    if largest < b:
        return largest+1
    else:
        return b
    
def binning(matrix):
    
    matrix = lil_matrix(matrix)
    
    indices_nz = matrix.nonzero()
    
    for i,j in zip(indices_nz[0], indices_nz[1]):
        matrix[i,j] = hash_count(matrix[i,j], b) 
   
    matrix = csr_matrix(matrix)
    
    return matrix

def remove_unnecessary(matrix, user_hash, user_reverse, song_hash, song_reverse):
    
    eps = 5 #value used for cold start
    
    while (1):

        change = False

        #SONGS
        counts_by_column = np.bincount(matrix.indices) 
        
        indices_to_keep = (counts_by_column > eps).nonzero() 
        indices_to_remove = (counts_by_column <= eps).nonzero()

        #remove songs from assistant dictionaries
        for idx in indices_to_remove[0]:
            song = song_reverse[idx]
            if song in song_hash:
                song_hash.pop(song)
            if idx in song_reverse:
                song_reverse.pop(idx)

        #update new indices in the assistant dictionaries
        reverse_tmp = dict()
        for i, idx in enumerate(indices_to_keep[0]):
            song = song_reverse[idx]
            song_hash[song] = i
            reverse_tmp[i] = song
            if idx in song_reverse:
                song_reverse.pop(idx)
        song_reverse = reverse_tmp


        if (indices_to_keep[0].shape[0] != matrix.shape[1]):
            change = True
            matrix = matrix[:,indices_to_keep[0]] #update matrix(remove songs)

        #USERS  
        counts_by_row = np.diff(matrix.indptr)
        indices_to_keep = (counts_by_row > eps).nonzero()
        indices_to_remove = (counts_by_row <= eps).nonzero()

        #remove users from assistant dictionaries
        for idx in indices_to_remove[0]:
            user = user_reverse[idx]
            if user in user_hash:
                user_hash.pop(user)
            if idx in user_reverse:
                user_reverse.pop(idx)

        #update new indices in the assistant dictionaries
        reverse_tmp = dict()
        for i, idx in enumerate(indices_to_keep[0]):
            user = user_reverse[idx]
            user_hash[user] = i
            reverse_tmp[i] = user
            if idx in user_reverse:
                user_reverse.pop(idx)
        user_reverse = reverse_tmp

        if (indices_to_keep[0].shape[0] != matrix.shape[0]):
            change = True
            matrix = matrix[indices_to_keep]    #update matrix(remove users)

        if change == False:
            break
        
    return matrix

def create_sparse_matrix(path = "train_triplets.txt", n = 300000, b = 10):

    #read data
    data_tmp = pd.read_table(path, 
                             usecols=[0, 1, 2], 
                             names=['user', 'song', 'plays'],
                             nrows=n)

    data = data_tmp.as_matrix()  

    #arrays which are needed for the creation of coo_matrix     
    rows = []
    columns = []
    play_counts = []
    
    #user_hash/song_hash is a dictionary with keys that are strings and values that are indices of row/column in the matrix
    #user_reverse/song_reverse is a dictionary with keys that are indices of row/column in the matrix and values that are strings
    #both of them are needed because we will delete several rows/columns in the matrix
    user_hash = dict()
    user_reverse = dict()
    song_hash = dict()
    song_reverse = dict()
    
    
    #iterate trough the datasets and create arrays: rows, columns, play_counts
    for i in range(n):
        u = data[i][0]
        s = data[i][1]
        
        if u not in user_hash:
            idx = len(user_hash)
            user_hash[u] = idx
            user_reverse[idx] = u
            
        if s not in song_hash:
            idx = len(song_hash)
            song_hash[s] = idx
            song_reverse[idx] = s
        
        rows.append(user_hash[u])
        columns.append(song_hash[s])
        
        play_counts.append(data[i][2])
        
    #create coo_matrix (duplicates user-songs will be automatically sumed up)
    matrix = coo_matrix( (play_counts, (rows,columns)), shape=(len(user_hash), len(song_hash)))
    #convert coo_matrix to crs_matrix 
    matrix = matrix.tocsr()
    
    #cold_start issue
    matrix = remove_unnecessary(matrix, user_hash, user_reverse, song_hash, song_reverse) 
    
    #binning
    matrix = binning(matrix)
    
    return matrix

def create_test_set(matrix, sample_size):

    M = lil_matrix(matrix)
    rows, cols = M.nonzero() 
    
    data_points = list(zip(rows, cols))
    random.shuffle(data_points)

    test_set = set()

    for i,j in data_points:
        test_set.add((i, j, M[i,j]))
        M[i,j] = 0
        
    return test_set

# This function is used for the calculation of root-mean-square error (RMSE). \n",
#The RMSE represents the sample standard deviation of the differences between predicted values and real(observed) values. \n",
#RMSE is calculated fot the dataset which consists of triples (row, column, value)\n",
def calculate_rmse(P, Q, test_set):

    prediction_matrix = np.dot(P.transpose(), Q)

    numerator = 0

    for triple in test_set:
        real_value = triple[2]
        predicted_value =  prediction_matrix[triple[0]][triple[1]]
        numerator += (predicted_value - real_value)**2

    return math.sqrt(numerator / len(test_set))

def randomSample(data, k):
    if not 0<=k<=len(data):
        raise ValueError('Required that 0 <= sample_size <= population_size')
    
    numbers = []
    numbersPicked = 0
    for i,number in enumerate(data):
        prob = (k-numbersPicked)/(len(data)-i)
        if random.random() < prob:
            numbers.append(number)
            numbersPicked += 1
    return numbers
```

# Read data
- Suggested size is 100.000 to have short enough execution times to try different variations.
- Gradient decent becomes very slow past 200.000
- Stochastic gradient decent becomes slow past 500.000

```python
# READ DATA, CREATE SPARSE MATRIX, CREATE TEST SET

start_time = time.time()
path = input("What is the path to the dataset? ")
n = int(input("How large should the subset be? "))
test_set_size = int(input("How large should the test set be? "))

b = 10
k = 30
#path = "train_triplets.txt"
#path = "C:\\Users\\Maki\\Desktop\\MMDS\\train_triplets\\train_triplets.txt"
#path = "C:\\Users\\Rene\\Documents\\GitLab\\LSH-Duplicate-Song-Search\\Task2\\train_triplets.txt"

matrix = create_sparse_matrix(path, n, b)

test_set = create_test_set(matrix, test_set_size)

print("Time: ", round(time.time()- start_time, 3))
print("Matrix shape: ", matrix.shape[0], matrix.shape[1])
```

# Gradient Decent Code

```python
matrix = lil_matrix(matrix)

def batchgradient(lmbda, learning_rate, iterations):
    start_time = time.time()
    k=30
    rows,cols = matrix.shape
    P = np.random.rand(k,rows)
    Q = np.random.rand(k,cols)

    u_indices,s_indices = matrix.nonzero()
    data_points = zip(u_indices, s_indices)
    m = len(u_indices)
    losshistory = []

    for q in range(iterations):
        error = 0
        data_points = zip(u_indices, s_indices)
        for uu,ii in data_points:
            error += (matrix[uu, ii] - np.dot((P[:,uu]).transpose(),Q[:,ii]) )**2
        #print('error: ', error/m)
        losshistory.append(error/m)


        data_points = zip(u_indices, s_indices)
        for u,i in data_points:

            e = matrix[u, i] - np.dot((P[:,u]).transpose(),Q[:,i]) 

            tmp = P[:,u]
            P[:,u] += learning_rate * ( 2* e * Q[:,i] - 2 * lmbda * P[:,u]) 
            Q[:,i] += learning_rate * ( 2* e * tmp - 2 * lmbda * Q[:,i]) 


    rmse = calculate_rmse(P, Q, test_set)
    print("Lambda:\t\t",lmbda,"\nLearning rate:\t", learning_rate, "\nIterations:\t",iterations)
    print("Time:\t\t",round(time.time()- start_time, 3))
    print('RMSE:\t\t', rmse)
    plt.plot(losshistory)
    plt.show()
```

# Stochastic Gradient Decent Code

```python
matrix = lil_matrix(matrix)

def stochasticgradient(lmbda, learning_rate, batch_size, iterations):
    start_time = time.time()
    rows,cols = matrix.shape
    P = np.random.rand(k,rows)
    Q = np.random.rand(k,cols)

    u_indices,s_indices = matrix.nonzero()
    data_points = zip(u_indices, s_indices)
    m = len(u_indices)
    nz = list(data_points)
    
    losshistory = []

    for q in range(iterations):
        error = 0
        subset = randomSample(nz, batch_size)

        for uu,ii in subset:
            error += (matrix[uu, ii] - np.dot((P[:,uu]).transpose(),Q[:,ii]) )**2
        #print('error: ', error/m)
        losshistory.append(error/m)

        for u,i in subset:
            e = matrix[u, i] - np.dot((P[:,u]).transpose(),Q[:,i]) 

            tmp = P[:,u]
            P[:,u] += learning_rate * ( 2* e * Q[:,i] - 2 * lmbda * P[:,u]) 
            Q[:,i] += learning_rate * ( 2* e * tmp - 2 * lmbda * Q[:,i]) 


    rmse = calculate_rmse(P, Q, test_set)
    print("Lambda:\t\t",lmbda,"\nLearning rate:\t", learning_rate, "\nBatch size:\t",batch_size,"\nIterations:\t",iterations)
    print("Time:\t\t",round(time.time()- start_time, 3))
    print('RMSE:\t\t', rmse)
    plt.plot(losshistory)
    plt.show()
```

```python
lmbda = 0.0001
learning_rate = 0.01
iterations = 3
batchgradient(lmbda, learning_rate, iterations)
```
